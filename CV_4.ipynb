```python
"""
Novel Hybrid Architecture for Lung Cancer Detection - Kaggle Notebook Version
===============================================================================

MHSA-CAGNet: Multi-Head Self-Attention with Cross-scale Attention Gating Network

Run each cell sequentially in your Kaggle notebook
"""

# ==============================================================================
# CELL 1: Install and Import Dependencies
# ==============================================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import numpy as np
from pathlib import Path
from PIL import Image
import cv2
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
import os
warnings.filterwarnings('ignore')

print("✓ All dependencies imported successfully")
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Available: {torch.cuda.is_available()}")


# ==============================================================================
# CELL 2: Configuration
# ==============================================================================

class Config:
    """Centralized configuration - Easy to modify"""
    # Paths
    DATA_ROOT = "/kaggle/input/iqothnccd-lung-cancer-dataset/The IQ-OTHNCCD lung cancer dataset/The IQ-OTHNCCD lung cancer dataset"
    OUTPUT_DIR = "/kaggle/working"
    
    # Training Hyperparameters
    IMG_SIZE = 224
    BATCH_SIZE = 16
    EPOCHS = 50
    LEARNING_RATE = 0.0001
    WEIGHT_DECAY = 0.01
    
    # Model Hyperparameters
    NUM_HEADS = 8
    DROPOUT_RATE = 0.5
    REDUCTION_RATIO = 16
    
    # Training Strategy
    N_FOLDS = 5
    RANDOM_SEED = 42
    NUM_WORKERS = 2
    
    # Advanced Options
    USE_MIXED_PRECISION = True
    SAVE_BEST_ONLY = True
    EARLY_STOPPING_PATIENCE = 10
    
    # Uncertainty Quantification
    MC_SAMPLES = 10  # Monte Carlo dropout samples
    
    @staticmethod
    def print_config():
        print("\n" + "="*60)
        print("CONFIGURATION")
        print("="*60)
        for key, value in vars(Config).items():
            if not key.startswith('_') and not callable(value):
                print(f"{key:30s}: {value}")
        print("="*60 + "\n")

Config.print_config()


# ==============================================================================
# CELL 3: Novel Component 1 - Frequency Domain Extractor
# ==============================================================================

class FrequencyDomainExtractor(nn.Module):
    """
    Novel Component 1: Frequency Domain Feature Extraction
    Uses FFT to capture texture patterns invisible in spatial domain
    
    Research Contribution: Dual-stream processing for enhanced feature learning
    """
    def __init__(self, in_channels=3):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, 64, 3, padding=1)
        self.conv2 = nn.Conv2d(64, 128, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        self.bn2 = nn.BatchNorm2d(128)
        self.relu = nn.ReLU(inplace=True)
        
    def forward(self, x):
        # Apply 2D FFT
        x_fft = torch.fft.rfft2(x, dim=(-2, -1))
        x_fft = torch.abs(x_fft)  # Magnitude spectrum
        
        # Process frequency features
        x = self.relu(self.bn1(self.conv1(x_fft)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

print("✓ Frequency Domain Extractor defined")


# ==============================================================================
# CELL 4: Novel Component 2 - Cross-Scale Attention Gate
# ==============================================================================

class CrossScaleAttentionGate(nn.Module):
    """
    Novel Component 2: Cross-Scale Attention Gating Mechanism
    
    Research Contribution: Dynamic feature weighting across different scales
    - Learns which features are important at each scale
    - Improves gradient flow in deep networks
    """
    def __init__(self, channels_high, channels_low, channels_int):
        super().__init__()
        self.W_gate = nn.Sequential(
            nn.Conv2d(channels_high, channels_int, 1),
            nn.BatchNorm2d(channels_int)
        )
        self.W_x = nn.Sequential(
            nn.Conv2d(channels_low, channels_int, 1),
            nn.BatchNorm2d(channels_int)
        )
        self.psi = nn.Sequential(
            nn.Conv2d(channels_int, 1, 1),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )
        self.relu = nn.ReLU(inplace=True)
        
    def forward(self, gate, skip):
        """
        Args:
            gate: Higher level features (coarse)
            skip: Lower level features (fine)
        Returns:
            Attention-weighted skip connection
        """
        g1 = self.W_gate(gate)
        x1 = self.W_x(skip)
        
        # Upsample gate to match skip connection size
        g1 = F.interpolate(g1, size=x1.shape[2:], mode='bilinear', align_corners=True)
        
        # Additive attention
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        
        return skip * psi

print("✓ Cross-Scale Attention Gate defined")


# ==============================================================================
# CELL 5: Novel Component 3 - Adaptive Feature Recalibration
# ==============================================================================

class AdaptiveFeatureRecalibration(nn.Module):
    """
    Novel Component 3: Adaptive Feature Recalibration (AFR)
    
    Research Contribution: Learnable fusion of channel and spatial attention
    - Unlike fixed fusion, learns optimal weighting during training
    - Adapts to different feature levels automatically
    """
    def __init__(self, channels, reduction=16):
        super().__init__()
        
        # Channel Attention Branch
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channels, channels // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channels // reduction, channels, bias=False),
            nn.Sigmoid()
        )
        
        # Spatial Attention Branch
        self.spatial_conv = nn.Sequential(
            nn.Conv2d(2, 1, 7, padding=3, bias=False),
            nn.Sigmoid()
        )
        
        # NOVEL: Learnable fusion weight (research contribution)
        self.alpha = nn.Parameter(torch.tensor(0.5))
        
    def forward(self, x):
        b, c, h, w = x.size()
        
        # Channel Attention
        avg_pool = self.avg_pool(x).view(b, c)
        max_pool = self.max_pool(x).view(b, c)
        channel_att = self.fc(avg_pool) + self.fc(max_pool)
        channel_att = channel_att.view(b, c, 1, 1)
        x_channel = x * channel_att
        
        # Spatial Attention
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        spatial_att = torch.cat([avg_out, max_out], dim=1)
        spatial_att = self.spatial_conv(spatial_att)
        x_spatial = x * spatial_att
        
        # Adaptive fusion with learnable weight
        alpha_sigmoid = torch.sigmoid(self.alpha)
        out = alpha_sigmoid * x_channel + (1 - alpha_sigmoid) * x_spatial
        
        return out

print("✓ Adaptive Feature Recalibration defined")


# ==============================================================================
# CELL 6: Novel Component 4 - Multi-Head Self-Attention
# ==============================================================================

class MultiHeadSelfAttention(nn.Module):
    """
    Novel Component 4: Multi-Head Self-Attention with Learned Positional Encoding
    
    Research Contribution: Learned positional embeddings for medical images
    - Captures long-range spatial dependencies
    - Positional encoding adapts to CT scan structure
    """
    def __init__(self, dim, num_heads=8, max_seq_len=196):
        super().__init__()
        self.num_heads = num_heads
        self.dim = dim
        self.head_dim = dim // num_heads
        
        assert dim % num_heads == 0, "dim must be divisible by num_heads"
        
        self.qkv = nn.Linear(dim, dim * 3)
        self.proj = nn.Linear(dim, dim)
        self.scale = self.head_dim ** -0.5
        
        # NOVEL: Learnable positional encoding
        self.pos_embedding = nn.Parameter(torch.randn(1, max_seq_len, dim) * 0.02)
        self.dropout = nn.Dropout(0.1)
        
    def forward(self, x):
        B, N, C = x.shape
        
        # Add learnable positional encoding
        if N <= self.pos_embedding.shape[1]:
            x = x + self.pos_embedding[:, :N, :]
        
        # Multi-head attention
        qkv = self.qkv(x).reshape(B, N, 3, self.num_heads, self.head_dim).permute(2, 0, 3, 1, 4)
        q, k, v = qkv[0], qkv[1], qkv[2]
        
        attn = (q @ k.transpose(-2, -1)) * self.scale
        attn = attn.softmax(dim=-1)
        attn = self.dropout(attn)
        
        x = (attn @ v).transpose(1, 2).reshape(B, N, C)
        x = self.proj(x)
        x = self.dropout(x)
        
        return x, attn

print("✓ Multi-Head Self-Attention defined")


#==============================================================================
# CELL 7: Novel Component 5 - Uncertainty-Aware Classification Head
# ==============================================================================

class UncertaintyAwareHead(nn.Module):
    """
    Novel Component 5: Uncertainty-Aware Classification with Monte Carlo Dropout
    
    Research Contribution: Quantifies prediction uncertainty for clinical trust
    - Essential for medical AI deployment
    - Helps identify ambiguous cases requiring expert review
    """
    def __init__(self, in_features, num_classes=1, dropout_rate=0.5):
        super().__init__()
        self.fc1 = nn.Linear(in_features, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, num_classes)
        self.dropout = nn.Dropout(dropout_rate)
        self.relu = nn.ReLU(inplace=True)
        
    def forward(self, x, training=True, mc_samples=10):
        """
        Args:
            x: Input features
            training: If True, use dropout; if False, estimate uncertainty
            mc_samples: Number of Monte Carlo samples for uncertainty
        """
        if training:
            # Standard forward pass with dropout
            x = self.relu(self.fc1(x))
            x = self.dropout(x)
            x = self.relu(self.fc2(x))
            x = self.dropout(x)
            x = self.fc3(x)
            return x
        else:
            # Monte Carlo Dropout for uncertainty estimation
            original_mode = self.training
            self.train()  # Enable dropout
            
            predictions = []
            for _ in range(mc_samples):
                with torch.no_grad():
                    x_temp = self.relu(self.fc1(x))
                    x_temp = self.dropout(x_temp)
                    x_temp = self.relu(self.fc2(x_temp))
                    x_temp = self.dropout(x_temp)
                    x_temp = torch.sigmoid(self.fc3(x_temp))
                    predictions.append(x_temp)
            
            predictions = torch.stack(predictions)
            mean_pred = predictions.mean(dim=0)
            uncertainty = predictions.std(dim=0)
            
            # Restore original mode
            if not original_mode:
                self.eval()
            
            return mean_pred, uncertainty

print("✓ Uncertainty-Aware Classification Head defined")




# ==============================================================================
# CELL 8: Main Model Architecture - MHSA-CAGNet
# ==============================================================================

class MHSACAGNet(nn.Module):
    """
    MHSA-CAGNet: Multi-Head Self-Attention with Cross-scale Attention Gating Network
    
    Complete Novel Architecture combining all 5 novel components:
    1. Dual-stream processing (spatial + frequency)
    2. Cross-scale attention gates
    3. Adaptive feature recalibration
    4. Multi-head self-attention with learned positional encoding
    5. Uncertainty quantification
    
    Research Contributions:
    - First architecture to combine frequency domain + attention gates
    - Learnable fusion weights in feature recalibration
    - Uncertainty estimation for medical AI trustworthiness
    """
    def __init__(self, num_classes=1, num_heads=8, dropout_rate=0.5):
        super().__init__()
        
        # === Spatial Stream (Traditional CNN Path) ===
        self.spatial_conv1 = nn.Sequential(
            nn.Conv2d(3, 64, 7, stride=2, padding=3),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2, padding=1)
        )
        
        self.spatial_conv2 = self._make_layer(64, 128, num_blocks=2)
        self.spatial_conv3 = self._make_layer(128, 256, num_blocks=2, stride=2)
        self.spatial_conv4 = self._make_layer(256, 512, num_blocks=2, stride=2)
        
        # === Frequency Stream ===
        self.freq_stream = FrequencyDomainExtractor(3)
        self.freq_fusion = nn.Conv2d(128, 64, 1)
        
        # === Cross-Scale Attention Gates ===
        self.cag1 = CrossScaleAttentionGate(512, 256, 256)
        self.cag2 = CrossScaleAttentionGate(256, 128, 128)
        
        # === Adaptive Feature Recalibration Modules ===
        self.afr1 = AdaptiveFeatureRecalibration(512)
        self.afr2 = AdaptiveFeatureRecalibration(256)
        self.afr3 = AdaptiveFeatureRecalibration(128)
        
        # === Multi-Head Self-Attention ===
        self.mhsa = MultiHeadSelfAttention(512, num_heads=num_heads)
        
        # === Global Pooling ===
        self.gap = nn.AdaptiveAvgPool2d(1)
        self.gmp = nn.AdaptiveMaxPool2d(1)
        
        # === Uncertainty-Aware Classification ===
        self.classifier = UncertaintyAwareHead(1024, num_classes, dropout_rate)
        
    def _make_layer(self, in_channels, out_channels, num_blocks, stride=1):
        """Helper function to create conv blocks"""
        layers = []
        layers.append(nn.Conv2d(in_channels, out_channels, 3, stride=stride, padding=1))
        layers.append(nn.BatchNorm2d(out_channels))
        layers.append(nn.ReLU(inplace=True))
        
        for _ in range(num_blocks - 1):
            layers.append(nn.Conv2d(out_channels, out_channels, 3, padding=1))
            layers.append(nn.BatchNorm2d(out_channels))
            layers.append(nn.ReLU(inplace=True))
        
        return nn.Sequential(*layers)
        
    def forward(self, x, return_attention=False, return_uncertainty=False):
        # === Dual Stream Processing ===
        # Spatial Stream
        s1 = self.spatial_conv1(x)      # 64 x 56 x 56
        s2 = self.spatial_conv2(s1)     # 128 x 56 x 56
        s3 = self.spatial_conv3(s2)     # 256 x 28 x 28
        s4 = self.spatial_conv4(s3)     # 512 x 14 x 14
        
        # Frequency Stream
        freq = self.freq_stream(x)
        freq = self.freq_fusion(freq)
        
        # Fuse frequency with spatial
        freq_resized = F.interpolate(freq, size=s1.shape[2:], mode='bilinear', align_corners=True)
        s1 = s1 + freq_resized
        
        # === Apply Adaptive Feature Recalibration ===
        s4 = self.afr1(s4)
        s3 = self.afr2(s3)
        s2 = self.afr3(s2)
        
        # === Cross-Scale Attention Gating ===
        s3_gated = self.cag1(s4, s3)
        s2_gated = self.cag2(s3_gated, s2)
        
        # === Multi-Head Self-Attention ===
        B, C, H, W = s4.shape
        s4_flat = s4.flatten(2).transpose(1, 2)  # B x (HW) x C
        
        # Ensure fixed sequence length for positional encoding
        target_len = 196  # 14x14
        if s4_flat.shape[1] != target_len:
            s4_flat = F.adaptive_avg_pool1d(s4_flat.transpose(1, 2), target_len).transpose(1, 2)
        
        s4_attn, attn_weights = self.mhsa(s4_flat)
        s4_attn = s4_attn.transpose(1, 2).reshape(B, C, 14, 14)
        
        # === Global Pooling ===
        gap = self.gap(s4_attn).flatten(1)
        gmp = self.gmp(s4_attn).flatten(1)
        features = torch.cat([gap, gmp], dim=1)
        
        # === Classification ===
        if return_uncertainty and not self.training:
            mean_pred, uncertainty = self.classifier(features, training=False, mc_samples=Config.MC_SAMPLES)
            if return_attention:
                return mean_pred, uncertainty, attn_weights
            return mean_pred, uncertainty
        else:
            output = self.classifier(features, training=True)
            if return_attention:
                return output, attn_weights
            return output

print("✓ MHSA-CAGNet architecture defined")
print(f"  - Spatial stream depth: 4 blocks")
print(f"  - Frequency stream: FFT-based")
print(f"  - Attention gates: 2 levels")
print(f"  - Feature recalibration: 3 levels")
print(f"  - Self-attention heads: {Config.NUM_HEADS}")


# ==============================================================================
# CELL 9: Dataset Class
# ==============================================================================

class LungCancerDataset(Dataset):
    """
    Custom Dataset for IQ-OTH/NCCD Lung Cancer Dataset
    Binary Classification: Malignant (1) vs Non-Malignant (0)
    """
    def __init__(self, root_dir, transform=None):
        self.root_dir = Path(root_dir)
        self.transform = transform
        self.samples = []
        self.labels = []
        
        # Define paths
        benign_path = self.root_dir / "Bengin cases"
        malignant_path = self.root_dir / "Malignant cases"
        normal_path = self.root_dir / "Normal cases"
        
        # Load Benign (label 0)
        if benign_path.exists():
            for img_path in benign_path.glob("*.png"):
                self.samples.append(str(img_path))
                self.labels.append(0)
            for img_path in benign_path.glob("*.jpg"):
                self.samples.append(str(img_path))
                self.labels.append(0)
        
        # Load Malignant (label 1)
        if malignant_path.exists():
            for img_path in malignant_path.glob("*.png"):
                self.samples.append(str(img_path))
                self.labels.append(1)
            for img_path in malignant_path.glob("*.jpg"):
                self.samples.append(str(img_path))
                self.labels.append(1)
        
        # Load Normal (label 0)
        if normal_path.exists():
            for img_path in normal_path.glob("*.png"):
                self.samples.append(str(img_path))
                self.labels.append(0)
            for img_path in normal_path.glob("*.jpg"):
                self.samples.append(str(img_path))
                self.labels.append(0)
        
        print(f"\n✓ Dataset loaded successfully")
        print(f"  Total images: {len(self.samples)}")
        print(f"  Malignant: {sum(self.labels)} ({sum(self.labels)/len(self.labels)*100:.1f}%)")
        print(f"  Non-Malignant: {len(self.labels) - sum(self.labels)} ({(len(self.labels)-sum(self.labels))/len(self.labels)*100:.1f}%)")
    
    def __len__(self):
        return len(self.samples)
    
    def __getitem__(self, idx):
        img_path = self.samples[idx]
        label = self.labels[idx]
        
        # Load image
        try:
            image = Image.open(img_path).convert('RGB')
        except:
            print(f"Error loading {img_path}")
            # Return a blank image if error
            image = Image.new('RGB', (Config.IMG_SIZE, Config.IMG_SIZE))
        
        if self.transform:
            image = self.transform(image)
        
        return image, label

print("✓ Dataset class defined")


# ==============================================================================
# CELL 10: Data Augmentation and Transforms
# ==============================================================================

def get_transforms():
    """Get training and validation transforms"""
    
    train_transform = transforms.Compose([
        transforms.Resize((Config.IMG_SIZE, Config.IMG_SIZE)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    val_transform = transforms.Compose([
        transforms.Resize((Config.IMG_SIZE, Config.IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    return train_transform, val_transform

print("✓ Data transforms defined")


# ==============================================================================
# CELL 11: Training Functions
# ==============================================================================

def train_epoch(model, loader, criterion, optimizer, device, scaler=None):
    """Train for one epoch"""
    model.train()
    running_loss = 0.0
    predictions = []
    targets = []
    
    pbar = tqdm(loader, desc='Training', leave=False)
    for images, labels in pbar:
        images = images.to(device)
        labels = labels.to(device).float().unsqueeze(1)
        
        optimizer.zero_grad()
        
        # Mixed precision training
        if Config.USE_MIXED_PRECISION and scaler is not None:
            with torch.cuda.amp.autocast():
                outputs = model(images)
                loss = criterion(outputs, labels)
            
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        
        running_loss += loss.item()
        preds = torch.sigmoid(outputs).detach().cpu().numpy()
        predictions.extend(preds)
        targets.extend(labels.cpu().numpy())
        
        pbar.set_postfix({'loss': f'{loss.item():.4f}'})
    
    epoch_loss = running_loss / len(loader)
    
    # Calculate metrics
    predictions = np.array(predictions).flatten()
    targets = np.array(targets).flatten()
    auc = roc_auc_score(targets, predictions) if len(np.unique(targets)) > 1 else 0.5
    
    return epoch_loss, auc, predictions, targets


def validate(model, loader, criterion, device):
    """Validate the model"""
    model.eval()
    running_loss = 0.0
    predictions = []
    targets = []
    
    with torch.no_grad():
        for images, labels in tqdm(loader, desc='Validation', leave=False):
            images = images.to(device)
            labels = labels.to(device).float().unsqueeze(1)
            
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item()
            preds = torch.sigmoid(outputs).cpu().numpy()
            predictions.extend(preds)
            targets.extend(labels.cpu().numpy())
    
    epoch_loss = running_loss / len(loader)
    
    # Calculate metrics
    predictions = np.array(predictions).flatten()
    targets = np.array(targets).flatten()
    auc = roc_auc_score(targets, predictions) if len(np.unique(targets)) > 1 else 0.5
    
    return epoch_loss, auc, predictions, targets

print("✓ Training functions defined")


# ==============================================================================
# CELL 12: Visualization Functions
# ==============================================================================

def plot_training_history(history, fold):
    """Plot training curves"""
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Loss curves
    axes[0].plot(history['train_loss'], label='Train Loss', linewidth=2)
    axes[0].plot(history['val_loss'], label='Val Loss', linewidth=2)
    axes[0].set_xlabel('Epoch', fontsize=12)
    axes[0].set_ylabel('Loss', fontsize=12)
    axes[0].set_title(f'Fold {fold} - Loss Curves', fontsize=14, fontweight='bold')
    axes[0].legend(fontsize=10)
    axes[0].grid(True, alpha=0.3)
    
    # AUC curves
    axes[1].plot(history['train_auc'], label='Train AUC', linewidth=2)
    axes[1].plot(history['val_auc'], label='Val AUC', linewidth=2)
    axes[1].set_xlabel('Epoch', fontsize=12)
    axes[1].set_ylabel('AUC', fontsize=12)
    axes[1].set_title(f'Fold {fold} - AUC Curves', fontsize=14, fontweight='bold')
    axes[1].legend(fontsize=10)
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{Config.OUTPUT_DIR}/fold{fold}_training_history.png', dpi=150, bbox_inches='tight')
    plt.show()


def plot_confusion_matrix(targets, predictions, fold, threshold=0.5):
    """Plot confusion matrix"""
    pred_labels = (predictions >= threshold).astype(int)
    cm = confusion_matrix(targets, pred_labels)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True,
                xticklabels=['Non-Malignant', 'Malignant'],
                yticklabels=['Non-Malignant', 'Malignant'])
    plt.title(f'Fold {fold} - Confusion Matrix', fontsize=14, fontweight='bold')
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()
    plt.savefig(f'{Config.OUTPUT_DIR}/fold{fold}_confusion_matrix.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    return cm


def plot_roc_curve(targets, predictions, fold):
    """Plot ROC curve"""
    fpr, tpr, thresholds = roc_curve(targets, predictions)
    auc = roc_auc_score(targets, predictions)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, linewidth=2, label=f'AUC = {auc:.4f}')
    plt.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random')
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title(f'Fold {fold} - ROC Curve', fontsize=14, fontweight='bold')
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'{Config.OUTPUT_DIR}/fold{fold}_roc_curve.png', dpi=150, bbox_inches='tight')
    plt.show()

print("✓ Visualization functions defined")


# ==============================================================================
# CELL 13: Main Training Loop
# ==============================================================================

def train_model():
    """Main training function with K-Fold Cross-Validation"""
    
    # Set device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"\n{'='*60}")
    print(f"Training on: {device}")
    print(f"{'='*60}\n")
    
    # Get transforms
    train_transform, val_transform = get_transforms()
    
    # Load full dataset
    full_dataset = LungCancerDataset(
        root_dir=Config.DATA_ROOT,
        transform=train_transform
    )
    
    # Prepare for K-Fold Cross-Validation
    skf = StratifiedKFold(n_splits=Config.N_FOLDS, shuffle=True, random_state=Config.RANDOM_SEED)
    fold_results = []
    
    # Training loop for each fold
    for fold, (train_idx, val_idx) in enumerate(skf.split(full_dataset.samples, full_dataset.labels)):
        print(f"\n{'='*60}")
        print(f"FOLD {fold + 1}/{Config.N_FOLDS}")
        print(f"{'='*60}\n")
        
        # Create data subsets
        train_dataset = torch.utils.data.Subset(full_dataset, train_idx)
        val_dataset_subset = torch.utils.data.Subset(full_dataset, val_idx)
        
        # Apply different transform for validation
        val_dataset_full = LungCancerDataset(root_dir=Config.DATA_ROOT, transform=val_transform)
        val_dataset = torch.utils.data.Subset(val_dataset_full, val_idx)
        
        # Create data loaders
        train_loader = DataLoader(
            train_dataset,
            batch_size=Config.BATCH_SIZE,
            shuffle=True,
            num_workers=Config.NUM_WORKERS,
            pin_memory=True
        )
        
        val_loader = DataLoader(
            val_dataset,
            batch_size=Config.BATCH_SIZE,
            shuffle=False,
            num_workers=Config.NUM_WORKERS,
            pin_memory=True
        )
        
        print(f"Train samples: {len(train_dataset)}")
        print(f"Val samples: {len(val_dataset)}\n")
        
        # Initialize model
        model = MHSACAGNet(
            num_classes=1,
            num_heads=Config.NUM_HEADS,
            dropout_rate=Config.DROPOUT_RATE
        ).to(device)
        
        # Loss function with class weighting
        criterion = nn.BCEWithLogitsLoss()
        
        # Optimizer
        optimizer = torch.optim.AdamW(
            model.parameters(),
            lr=Config.LEARNING_RATE,
            weight_decay=Config.WEIGHT_DECAY
        )
        
        # Learning rate scheduler
        scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(
            optimizer,
            T_0=10,
            T_mult=2,
            eta_min=1e-6
        )
        
        # Mixed precision scaler
        scaler = torch.cuda.amp.GradScaler() if Config.USE_MIXED_PRECISION else None
        
        # Training history
        history = {
            'train_loss': [],
            'val_loss': [],
            'train_auc': [],
            'val_auc': []
        }
        
        best_auc = 0.0
        patience_counter = 0
        
        # Training loop
        for epoch in range(Config.EPOCHS):
            print(f"\nEpoch {epoch+1}/{Config.EPOCHS}")
            print("-" * 40)
            
            # Train
            train_loss, train_auc, train_preds, train_targets = train_epoch(
                model, train_loader, criterion, optimizer, device, scaler
            )
            
            # Validate
            val_loss, val_auc, val_preds, val_targets = validate(
                model, val_loader, criterion, device
            )
            
            # Update scheduler
            scheduler.step()
            
            # Store history
            history['train_loss'].append(train_loss)
            history['val_loss'].append(val_loss)
            history['train_auc'].append(train_auc)
            history['val_auc'].append(val_auc)
            
            # Print metrics
            print(f"Train Loss: {train_loss:.4f} | Train AUC: {train_auc:.4f}")
            print(f"Val Loss:   {val_loss:.4f} | Val AUC:   {val_auc:.4f}")
            print(f"LR: {optimizer.param_groups[0]['lr']:.6f}")
            
            # Save best model
            if val_auc > best_auc:
                best_auc = val_auc
                patience_counter = 0
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': model.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'auc': best_auc,
                    'fold': fold + 1
                }, f'{Config.OUTPUT_DIR}/best_model_fold{fold+1}.pth')
                print(f"✓ Best model saved! AUC: {best_auc:.4f}")
            else:
                patience_counter += 1
            
            # Early stopping
            if patience_counter >= Config.EARLY_STOPPING_PATIENCE:
                print(f"\n⚠ Early stopping triggered at epoch {epoch+1}")
                break
        
        # Load best model for final evaluation
        checkpoint = torch.load(f'{Config.OUTPUT_DIR}/best_model_fold{fold+1}.pth', weights_only=False)
        model.load_state_dict(checkpoint['model_state_dict'])
        
        # Final validation
        _, final_auc, final_preds, final_targets = validate(
            model, val_loader, criterion, device
        )
        
        # Plot results
        plot_training_history(history, fold + 1)
        plot_confusion_matrix(final_targets, final_preds, fold + 1)
        plot_roc_curve(final_targets, final_preds, fold + 1)
        
        # Classification report
        pred_labels = (final_preds >= 0.5).astype(int)
        print(f"\n{classification_report(final_targets, pred_labels, target_names=['Non-Malignant', 'Malignant'])}")
        
        # Store fold results
        fold_results.append({
            'fold': fold + 1,
            'best_auc': best_auc,
            'final_auc': final_auc,
            'history': history
        })
    
    # Summary of all folds
    print(f"\n{'='*60}")
    print("CROSS-VALIDATION SUMMARY")
    print(f"{'='*60}")
    
    for result in fold_results:
        print(f"Fold {result['fold']}: Best AUC = {result['best_auc']:.4f} | Final AUC = {result['final_auc']:.4f}")
    
    all_aucs = [r['best_auc'] for r in fold_results]
    mean_auc = np.mean(all_aucs)
    std_auc = np.std(all_aucs)
    
    print(f"\n{'='*60}")
    print(f"Mean AUC: {mean_auc:.4f} ± {std_auc:.4f}")
    print(f"{'='*60}\n")
    
    # Plot overall results
    plt.figure(figsize=(10, 6))
    folds = [r['fold'] for r in fold_results]
    aucs = [r['best_auc'] for r in fold_results]
    
    plt.bar(folds, aucs, color='skyblue', edgecolor='navy', linewidth=2)
    plt.axhline(mean_auc, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_auc:.4f}')
    plt.xlabel('Fold', fontsize=12)
    plt.ylabel('Best AUC', fontsize=12)
    plt.title('Cross-Validation Results - AUC per Fold', fontsize=14, fontweight='bold')
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.savefig(f'{Config.OUTPUT_DIR}/cv_summary.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    return fold_results

print("✓ Main training function defined")

        




# ==============================================================================
# CELL 14: Uncertainty Estimation Function
# ==============================================================================

def estimate_uncertainty(model_path, data_loader, device):
    """
    Estimate prediction uncertainty using Monte Carlo Dropout
    Useful for identifying cases that need expert review
    """
    model = MHSACAGNet(
        num_classes=1,
        num_heads=Config.NUM_HEADS,
        dropout_rate=Config.DROPOUT_RATE
    ).to(device)
    
    checkpoint = torch.load(model_path, weights_only=False)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    all_predictions = []
    all_uncertainties = []
    all_targets = []
    
    print("\nEstimating prediction uncertainty...")
    with torch.no_grad():
        for images, labels in tqdm(data_loader):
            images = images.to(device)
            
            # Get predictions with uncertainty
            mean_pred, uncertainty = model(images, return_uncertainty=True)
            
            all_predictions.extend(mean_pred.cpu().numpy())
            all_uncertainties.extend(uncertainty.cpu().numpy())
            all_targets.extend(labels.numpy())
    
    all_predictions = np.array(all_predictions).flatten()
    all_uncertainties = np.array(all_uncertainties).flatten()
    all_targets = np.array(all_targets)
    
    # Visualize uncertainty
    plt.figure(figsize=(12, 5))
    
    # Histogram of uncertainties
    plt.subplot(1, 2, 1)
    plt.hist(all_uncertainties, bins=50, color='steelblue', edgecolor='black', alpha=0.7)
    plt.xlabel('Uncertainty (Standard Deviation)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title('Distribution of Prediction Uncertainty', fontsize=14, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    # Scatter plot: Prediction vs Uncertainty
    plt.subplot(1, 2, 2)
    colors = ['green' if t == 0 else 'red' for t in all_targets]
    plt.scatter(all_predictions, all_uncertainties, c=colors, alpha=0.5, s=20)
    plt.xlabel('Predicted Probability', fontsize=12)
    plt.ylabel('Uncertainty', fontsize=12)
    plt.title('Prediction vs Uncertainty', fontsize=14, fontweight='bold')
    plt.legend(['Non-Malignant', 'Malignant'], fontsize=10)
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{Config.OUTPUT_DIR}/uncertainty_analysis.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    # Identify high uncertainty cases
    high_uncertainty_threshold = np.percentile(all_uncertainties, 90)
    high_uncertainty_indices = np.where(all_uncertainties > high_uncertainty_threshold)[0]
    
    print(f"\nUncertainty Analysis:")
    print(f"  Mean uncertainty: {np.mean(all_uncertainties):.4f}")
    print(f"  Std uncertainty: {np.std(all_uncertainties):.4f}")
    print(f"  High uncertainty cases (>90th percentile): {len(high_uncertainty_indices)}")
    print(f"  These cases may require expert review for verification")
    
    return all_predictions, all_uncertainties, all_targets

print("✓ Uncertainty estimation function defined")


# ==============================================================================
# CELL 15: Attention Visualization
# ==============================================================================

def visualize_attention(model_path, image_tensor, device, save_path=None):
    """
    Visualize attention maps to understand what the model focuses on
    """
    model = MHSACAGNet(
        num_classes=1,
        num_heads=Config.NUM_HEADS,
        dropout_rate=Config.DROPOUT_RATE
    ).to(device)
    
    checkpoint = torch.load(model_path, weights_only=False)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    with torch.no_grad():
        image_tensor = image_tensor.unsqueeze(0).to(device)
        output, attn_weights = model(image_tensor, return_attention=True)
        pred = torch.sigmoid(output).item()
    
    # Average attention across heads
    attn_weights = attn_weights.cpu().numpy()
    attn_avg = attn_weights.mean(axis=1)[0]  # Average over heads
    
    # Reshape to spatial dimensions
    attn_map = attn_avg.mean(axis=0).reshape(14, 14)
    
    # Resize to match input
    attn_map_resized = cv2.resize(attn_map, (Config.IMG_SIZE, Config.IMG_SIZE))
    
    # Denormalize image
    image_np = image_tensor.cpu().squeeze(0).permute(1, 2, 0).numpy()
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    image_np = std * image_np + mean
    image_np = np.clip(image_np, 0, 1)
    
    # Plot
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Original image
    axes[0].imshow(image_np)
    axes[0].set_title('Original Image', fontsize=12, fontweight='bold')
    axes[0].axis('off')
    
    # Attention map
    im = axes[1].imshow(attn_map_resized, cmap='jet')
    axes[1].set_title('Attention Map', fontsize=12, fontweight='bold')
    axes[1].axis('off')
    plt.colorbar(im, ax=axes[1], fraction=0.046, pad=0.04)
    
    # Overlay
    axes[2].imshow(image_np)
    axes[2].imshow(attn_map_resized, cmap='jet', alpha=0.5)
    axes[2].set_title(f'Overlay (Pred: {pred:.3f})', fontsize=12, fontweight='bold')
    axes[2].axis('off')
    
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()
    
    return attn_map_resized, pred

print("✓ Attention visualization function defined")



# ==============================================================================
# CELL 16: Model Summary
# ==============================================================================

def print_model_summary():
    """Print detailed model architecture summary"""
    model = MHSACAGNet(num_classes=1)
    
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    
    print(f"\n{'='*60}")
    print("MODEL ARCHITECTURE SUMMARY")
    print(f"{'='*60}")
    print(f"Model Name: MHSA-CAGNet")
    print(f"Total Parameters: {total_params:,}")
    print(f"Trainable Parameters: {trainable_params:,}")
    print(f"{'='*60}")
    
    print("\nNovel Components:")
    print("  1. ✓ Frequency Domain Extractor (FFT-based)")
    print("  2. ✓ Cross-Scale Attention Gates (2 levels)")
    print("  3. ✓ Adaptive Feature Recalibration (learnable fusion)")
    print("  4. ✓ Multi-Head Self-Attention (learned positional encoding)")
    print("  5. ✓ Uncertainty-Aware Classification Head")
    
    print("\nArchitecture Highlights:")
    print("  • Dual-stream processing (spatial + frequency)")
    print("  • 4-level spatial encoder")
    print("  • 3-level feature recalibration")
    print("  • 8-head self-attention mechanism")
    print("  • Monte Carlo dropout uncertainty")
    print(f"{'='*60}\n")

print_model_summary()


# ==============================================================================
# CELL 17: RUN TRAINING
# ==============================================================================
# Uncomment the line below to start training

fold_results = train_model()


# ==============================================================================
# CELL 18: EXAMPLE - Uncertainty Estimation (Run after training)
# ==============================================================================
"""
# Example: Estimate uncertainty on validation set from fold 1
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
_, val_transform = get_transforms()

val_dataset = LungCancerDataset(
    root_dir=Config.DATA_ROOT,
    transform=val_transform
)

val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

predictions, uncertainties, targets = estimate_uncertainty(
    model_path=f'{Config.OUTPUT_DIR}/best_model_fold1.pth',
    data_loader=val_loader,
    device=device
)
"""


# ==============================================================================
# CELL 19: EXAMPLE - Visualize Attention (Run after training)
# ==============================================================================
"""
# Example: Visualize attention for a single image
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
_, val_transform = get_transforms()

# Load a sample image
sample_dataset = LungCancerDataset(
    root_dir=Config.DATA_ROOT,
    transform=val_transform
)

# Get first image
image_tensor, label = sample_dataset[0]

attn_map, pred = visualize_attention(
    model_path=f'{Config.OUTPUT_DIR}/best_model_fold1.pth',
    image_tensor=image_tensor,
    device=device,
    save_path=f'{Config.OUTPUT_DIR}/attention_example.png'
)

print(f"True label: {label}, Predicted: {pred:.4f}")
"""


# ==============================================================================
# CELL 20: Test Set Inference (If you have test cases)
# ==============================================================================
"""
def predict_test_set(model_path, test_dir, device):
    '''
    Make predictions on unlabeled test set
    '''
    _, val_transform = get_transforms()
    
    # Load model
    model = MHSACAGNet(num_classes=1).to(device)
    checkpoint = torch.load(model_path)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    # Get test images
    test_images = list(Path(test_dir).glob('*.png')) + list(Path(test_dir).glob('*.jpg'))
    
    results = []
    
    print(f"Predicting on {len(test_images)} test images...")
    
    for img_path in tqdm(test_images):
        # Load and preprocess
        image = Image.open(img_path).convert('RGB')
        image_tensor = val_transform(image).unsqueeze(0).to(device)
        
        with torch.no_grad():
            # Get prediction with uncertainty
            mean_pred, uncertainty = model(image_tensor, return_uncertainty=True)
            pred_prob = mean_pred.item()
            unc = uncertainty.item()
        
        results.append({
            'filename': img_path.name,
            'prediction_probability': pred_prob,
            'predicted_class': 'Malignant' if pred_prob >= 0.5 else 'Non-Malignant',
            'uncertainty': unc,
            'confidence': 'High' if unc < 0.1 else 'Medium' if unc < 0.2 else 'Low'
        })
    
    # Save results
    import pandas as pd
    df_results = pd.DataFrame(results)
    df_results.to_csv(f'{Config.OUTPUT_DIR}/test_predictions.csv', index=False)
    
    print(f"✓ Predictions saved to {Config.OUTPUT_DIR}/test_predictions.csv")
    
    return df_results

# Example usage:
# test_predictions = predict_test_set(
#     model_path=f'{Config.OUTPUT_DIR}/best_model_fold1.pth',
#     test_dir='/kaggle/input/your-dataset/Test cases',
#     device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# )
"""

print("\n" + "="*60)
print("ALL CELLS DEFINED SUCCESSFULLY!")
print("="*60)
print("\n📋 INSTRUCTIONS:")
print("1. Verify Config.DATA_ROOT path matches your dataset location")
print("2. Run Cell 17 to start training: fold_results = train_model()")
print("3. After training, use Cell 18 for uncertainty analysis")
print("4. Use Cell 19 to visualize attention maps")
print("5. Use Cell 20 for test set predictions")
print("\n🚀 Ready to train your novel MHSA-CAGNet model!")
print("="*60)
```
